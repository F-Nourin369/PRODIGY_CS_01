```python
import tkinter as tk
from tkinter import messagebox

def encrypt(text, shift):
    encrypted_text = ""
    for char in text:
        if char.isalpha():
            shift_amount = shift % 26
            if char.islower():
                encrypted_text += chr((ord(char) - 97 + shift_amount) % 26 + 97)
            elif char.isupper():
                encrypted_text += chr((ord(char) - 65 + shift_amount) % 26 + 65)
        else:
            encrypted_text += char
    return encrypted_text

def decrypt(text, shift):
    return encrypt(text, -shift)

def perform_action():
    text = message_entry.get("1.0", tk.END).strip()
    shift = shift_entry.get().strip()
    if not text or not shift.isdigit():
        messagebox.showerror("Input Error", "Please enter a valid message and shift value.")
        return
    shift = int(shift)
    if action_var.get() == 1:
        result = encrypt(text, shift)
        result_label.config(text=f"Encrypted message: {result}")
    else:
        result = decrypt(text, shift)
        result_label.config(text=f"Decrypted message: {result}")

app = tk.Tk()
app.title("Caesar Cipher Encryption and Decryption")

# Message input
tk.Label(app, text="Enter your message:").grid(row=0, column=0, padx=10, pady=10)
message_entry = tk.Text(app, height=5, width=40)
message_entry.grid(row=0, column=1, padx=10, pady=10)

# Shift input
tk.Label(app, text="Enter the shift value:").grid(row=1, column=0, padx=10, pady=10)
shift_entry = tk.Entry(app)
shift_entry.grid(row=1, column=1, padx=10, pady=10)

# Action selection
action_var = tk.IntVar()
tk.Radiobutton(app, text="Encrypt", variable=action_var, value=1).grid(row=2, column=0, padx=10, pady=10)
tk.Radiobutton(app, text="Decrypt", variable=action_var, value=2).grid(row=2, column=1, padx=10, pady=10)
action_var.set(1)

# Perform button
perform_button = tk.Button(app, text="Perform", command=perform_action)
perform_button.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

# Result label
result_label = tk.Label(app, text="")
result_label.grid(row=4, column=0, columnspan=2, padx=10, pady=10)

app.mainloop()
```
